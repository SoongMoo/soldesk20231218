```python
''' 집계함수 이외의 컬럼은 group by절에 있는 컬럼만 사용 가능
select year, min(lifeExp),  max(lifeExp), avg(lifeExp), sum(lifeExp), count(lifeExp)
from df
group by year
'''
df.groupby('year')['lifeExp'].min()
df.groupby('year').lifeExp.max()
df.groupby('year').lifeExp.mean()
df.groupby('year').lifeExp.sum()
```

```python
# 년도별 평균수명에 대한 평균
'''
select year, continent, min(lifeExp),  max(lifeExp), avg(lifeExp), sum(lifeExp)
      , count(lifeExp)
from df
group by year, continent
'''
df.groupby(['year','continent'])['lifeExp'].min()
df.groupby(['year','continent'])['lifeExp'].max()
df.groupby(['year','continent'])['lifeExp'].mean()
df.groupby(['year','continent'])['lifeExp'].sum()
```

```python
### 년도별 대륙의 평균수명에 대한 평균과 gdp의 평균을 구하세요?
'''
select year, continent, avg(lifeExp), avg(gdpPercap), min(lifeExp), min(gdpPercap)
from df
group by year, continent
'''
df.groupby(['year','continent'])[['lifeExp','gdpPercap']].mean()
df.groupby(['year','continent'])[['lifeExp','gdpPercap']].min()
```
